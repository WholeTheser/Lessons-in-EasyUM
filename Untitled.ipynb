```python
import random as rd
```

```python
def passwords1(level):
    digits = [*'1234567890']
    simbols = [*'!@#$%^&*/.-?,<>\|']
    letters = [*'qwertyuiopasdfghjklzxcvbnm']
    if level == 'low':
        return ''.join(rd.choices(letters, k=8))
    elif level == 'med':
        prepass = [i.upper() if rd.randint(0, 1) else i for i in rd.choices(letters, k=8)]+rd.choices(digits, k=4)
        rd.shuffle(prepass)
        return ''.join(prepass)
    else:
        prepass = [i.upper() if rd.randint(0, 1) else i for i in rd.choices(letters, k=8)] + \
        rd.choices(digits, k=8) + \
        rd.choices(simbols, k=4)
        rd.shuffle(prepass)
        return ''.join(prepass)
```

```python
for i in range(10):
    print(passwords1('qwerty'))
```

```python
def passwords2(length):
    digits = [*'1234567890']
    simbols = [*'!@#$%^&*/.-?,<>\|']
    letters = [i.upper() if rd.randint(0, 1) else i for i in [*'qwertyuiopasdfghjklzxcvbnm']]
    symbols = [letters, digits, simbols]
    if length >=8:
        my_symbols = symbols[:length//8+1]
        count = length//len(my_symbols)
        password_res = ''
        for massiv in my_symbols:
            password_res += ''.join(rd.choices(massiv, k=count))
        if len(password_res) < length:
            need = length - len(password_res)
            password_res += ''.join(rd.choices(rd.choice(my_symbols), k=need))
        return password_res
    else:
        print('Больше 8')
```

```python
passwords2(16)
```
